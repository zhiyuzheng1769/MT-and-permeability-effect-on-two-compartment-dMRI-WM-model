```python
from local_optim_fit import reorder_df, forge_axcaliber_250k, fit_dict_axcaliber, local_optim_test_indata
from dmipy.signal_models import gaussian_models, cylinder_models
from dmipy.core.acquisition_scheme import acquisition_scheme_from_qvalues
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from dmipy.distributions import distribute_models
from scipy.optimize import curve_fit, minimize
import pandas as pd
from scipy.stats import gamma
from dmipy.core.acquisition_scheme import acquisition_scheme_from_gradient_strengths
from scipy.interpolate import CubicSpline, interp1d
from matplotlib.ticker import LinearLocator
import nibabel as nib
import copy
import pickle
```

```python
costs = []
params = []
forward_model = forge_axcaliber_250k()
for perm in np.arange(0, 0.021, 0.001):
    
    simdata = pd.read_csv("./perm/signal_MT_0_sus_0_perm_" + '{0:.3f}'.format(perm) + "_rmean_2.00_density_0.65.csv", header=None)

    y = simdata[2].to_numpy()/simdata[2][0]
    cost, param = local_optim_test_indata(forward_model, y)
    costs.append(cost)
    params.append(param)
```

```python
file_path = "cost_distributed_perm_0-0.02_final.pickle"
with open(file_path, 'wb') as file:
    pickle.dump(costs, file)

file_path = "params_distributed_perm_0-0.02_final.pickle"
with open(file_path, 'wb') as file:
    pickle.dump(params, file)
```

```python
# perm_list = [0, 0.005, 0.01, 0.015, 0.02]
cf_cost = []
nm_cost = []
pl_cost = []
lb_cost = []
tnc_cost = []
cobyla_cost = []
slsqp_cost = []
tc_cost = []

for a in costs:
    cf_cost.append(a["curve_fit"])
    nm_cost.append(a["Nelder-Mead"])
    pl_cost.append(a["Powell"])
    lb_cost.append(a["L-BFGS-B"])
    tnc_cost.append(a["TNC"])
    cobyla_cost.append(a["COBYLA"])
    slsqp_cost.append(a['SLSQP'])
    tc_cost.append(a["trust-constr"])
    
plt.plot(np.arange(0, 0.021, 0.001), nm_cost)
# plt.plot(np.arange(0, 0.021, 0.001), nm_cost_sing)
plt.xlabel("permeability")
plt.ylabel("Estimation square error")
# plt.legend()
# # if perm is not None:
# #     plt.title("Permeability="+str(perm))
plt.show()
```
